```python
#Load the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data
X = []
Y = []
for line in open('data_1d.csv'):
    x, y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
```

```python
# let's turn X and Y into numpy arrays since that will be useful later(For vector-matrix ops)
X = np.array(X)
Y = np.array(Y)
```

```python
# let's plot the data to see what it looks like
plt.scatter(X, Y)
plt.show()
```

```python
#linear regression function 
# y = a*x + b
# we need tofigure out the value of regression coefficients
denominator = X.dot(X) - X.mean()*X.sum()
a = (X.dot(Y) - Y.mean()*X.sum())/denominator
b = (Y.mean()*X.dot(X) - X.mean()*X.dot(Y))/denominator
```

```python
print a
print b
```

```python
#let's predict yhat
Yhat = a*X+b
```

```python
#let's predict everything together to make sure it works
plt.scatter(X,Y)
plt.plot(X,Yhat)
plt.show()
```

```python
# determine how good the model is by computing the r-squared
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print("the r-squared is:", r2)
```
